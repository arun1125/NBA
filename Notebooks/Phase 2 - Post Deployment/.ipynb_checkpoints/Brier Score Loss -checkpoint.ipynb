```python
import pandas as pd
import numpy as np
from sklearn.metrics import brier_score_loss
```

```python
model_scores = pd.read_csv('../../Data/historical_pbp_modelled.csv').dropna()
```

```python
global_b_score = brier_score_loss(model_scores['home_team_win'], model_scores['preds_wO_elo'])
```

```python
b_scores = model_scores.apply(lambda x: brier_score_loss([x['home_team_win']],[x['preds_wO_elo']]) , axis =1)
```

```python
model_scores = model_scores.join(np.log(agg_games[('brier_score', 'mean')]).rename('mean_brier_scores'),
                 how='left', on='GAME_ID')
```

```python
np.percentile(model_scores['log_std_brier_scores'].unique(), 0.9)
```

```python
def sample_game(col, threshold=0.9):
    threshold = np.percentile(model_scores[col].unique(), threshold)
    return np.random.choice(model_scores.query(f"{col} > {threshold}")['GAME_ID'].unique())
```

```python
def plot(game_id):
    
    temp = model_scores.query(f"GAME_ID == {game_id}")
    
    temp['preds_w_elo'].plot(x = (temp['time_remaining']))
    
```

```python
game_id = sample_game('mean_brier_scores', 0.1)

game = model_scores.query(f'GAME_ID == {game_id}')['brier_score']

fig, ax = plt.subplots(figsize=(20,16))
for i in range(400, 0, -1):
    game.rolling(i).mean().plot(ax= ax)
    
plt.show()
```
